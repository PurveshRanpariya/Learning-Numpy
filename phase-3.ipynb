```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("==== Zomato sales analysis ==== ")
print("\n Sales data shape", sales_data.shape)
print("\n Sample data for 1st 3 restau: \n", sales_data[:3])
print("\n Sample data for 1st 3 restau: \n", sales_data[:, 1:]) #  ":" = all rows, "1:" = col 1 to end
```

```python
#total sale per year

column_sum = (np.sum(sales_data, axis=0)) # col wise sum

print(column_sum[1:])

yearly_total = np.sum(sales_data, axis=0)[1:] # excluding 1st col
years = [2021, 2022, 2023, 2024]
print("\n Yearly total sales from 2021 to 2024: ", yearly_total)

# Minimum sales in a year
min_sales = np.min(sales_data[:, 1:], axis=0)
print("\n Minimum sales per year from 2021 to 2024: ", min_sales)

# Maximun sales per restaurent
max_sale = np.max(sales_data[:, 1:], axis=1) # row wise max
print("\n Maximum sales per restaurent from 2021 to 2024: ", max_sale)

# Average sales per restaurent
avg_sale = np.mean(sales_data[:, 1:], axis=1) # row wise mean
print("\n Average sales per restaurent from 2021 to 2024: ", avg_sale)

total_sum = np.sum(column_sum, axis=0) 
print("\n Total sales from 2021 to 2024: ", total_sum)

```

```python
cumsum = np.cumsum(sales_data[:, 1:], axis=1) # cumulative sum
print("\n Cumulative sum of sales from 2021 to 2024: \n", cumsum)

plt.Figure(figsize=(10,6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Avrrage cumulative sales across restaurents")
plt.xlabel("Years from 2021 to 2024")
plt.ylabel("sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10])

print("Vector addition: ", vector1 + vector2)

print("\n vector multiplication: ", vector1 * vector2)

print("\n dot product: ", np.dot(vector1, vector2))

angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print("\n Angle between vectors (in radians): ", angle)
print("\n Angle between vectors (in degrees): ", np.degrees(angle))
```
